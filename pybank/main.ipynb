```python
csv_file_path = Path('./pybank/budget_data.csv')  

total_months = 1
total = 0

max_increase_date = ""
max_increase_amount = 0

max_decrease_date = ""
max_decrease_amount = 0

with open(csv_file_path, mode='r') as file:
   
    lines = file.readlines()

    data_lines = lines[1:]

    total_changes = 0
    num_changes = 0

    total = int(data_lines[0].strip().split(',')[1])

    for i in range(1, len(data_lines)):
        
        current_columns = data_lines[i].strip().split(',')
        previous_columns = data_lines[i - 1].strip().split(',')

        # Calculate the change in Profit/Losses
        change = int(current_columns[1]) - int(previous_columns[1])

        # Add the change to the total
        total_changes += change
        num_changes += 1


        if change > max_increase_amount:
            max_increase_amount = change
            max_increase_date = current_columns[0]

        if change < max_decrease_amount:
            max_decrease_amount = change
            max_decrease_date = current_columns[0]

        total_months+=1

       
        total += int(current_columns[1])


average_change = total_changes / num_changes if num_changes > 0 else 0

average_change = round(average_change,2)

# Print the total profit/loss
print(f'TOTAL MONTHS = {total_months}')
print(f'TOTAL: ${total}')
print(f'Average Change: ${average_change}')
print(f'Greatest increase in profits {max_increase_date} (${max_increase_amount})')
print(f'Greatest decrease in profits: {max_decrease_date} (${max_decrease_amount})')
```
